```python
import time
import requests
from bs4 import BeautifulSoup
```

```python
product_page_links = []
max_page = 1538
# max_page=10
for i in range(0, max_page):
    url = "https://cosmeet.cosme.net/product/search/page/{0}/srt/4/itm/1003".format(i)
    response = requests.get(url)
    bsobj = BeautifulSoup(response.content, "lxml")
    product_list = bsobj.find("div", {"id":"cnt-list"})
    for product in product_list.findAll("div", {"class", "info"}):
        product_page_links.append(product.find("a")['href'])
    time.sleep(2)
```

```python
import pandas as pd
pd.DataFrame(product_page_links).to_csv('./product_links.csv')
```

```python
import re
def get_item_info(bsobj):
    global DF_PRODUCTS
    try:
        item_name = bsobj.find("strong", {"class": "pdct-name fn"}).get_text()
        item_rate = bsobj.find("p", {"class": re.compile("average avr.*")}).get_text()
        item_point = bsobj.find("p", {"class": "point"}).get_text()
        item_amt_price = bsobj.findAll("li", {"class": "clearfix"})[2].find("p", {"class": "info-desc"}).get_text()
    #     item_date = bsobj.findAll("li", {"class": "clearfix"})[3].find("p", {"class": "info-desc"}).get_text()    
        DF_PRODUCTS = DF_PRODUCTS.append([[item_name, item_rate, item_point, item_amt_price]])
    except:
        print('error')
```

```python
def create_item_data(url_top):
    print(url_top)
    response = requests.get(url_top)
    bsobj = BeautifulSoup(response.content, "lxml")
    get_item_info(bsobj)
    time.sleep(2)
```
